````python
# Fazer o exercício aqui
class Produto:
    __codigo = 'privado'
    __nome = 'privado'
    __preco = 'privado'
    __tipo = 'privado'

    def __init__(self,codigo,nome,preco,tipo):
        self.__codigo = codigo
        self.__nome = nome
        self.__preco = preco
        self.__tipo = tipo


    @property
    def codigo (self):
        return self.__codigo
    @property
    def nome (self):
        return self.__nome
    @property
    def preco(self):
        return self.__preco 
    @property
    def tipo(self):
        return self.__tipo   

    
    @codigo.setter
    def codigo (self,novoCodigo):
        self.__codigo = novoCodigo
    @nome.setter
    def nome (self, novoNome):
        self.__nome = novoNome
    @preco.setter
    def preco(self, novoPreco):
        self.__preco = novoPreco
    @tipo.setter
    def tipo(self, novoTipo):
        self.__tipo = novoTipo


    def calculo_total(self):
        self.quantidade = float(input("Quantidade produtos desejados:"))
        self.valor = self.preco * self.quantidade

        return "Na compra de {} {} ira pagar {:.2f}€\n".format(int(self.quantidade),self.nome,self.valor)
        


    def __str__(self):
        return "Produto: {} || Preco: {:.2f}€ || Tipo: {} || Codigo: {}\n".format(
            self.nome,self.preco,self.tipo,self.codigo
            )


class Pedido(Produto):
    lista_produtos = []
    lista_quantidades = []

    def __init__(self,lista_produtos=[],lista_quantidade=[]):
        self.lista_produtos =lista_produtos
        self.lista_quantidades= lista_quantidade
        
    def mostrar_pedido(self):

        for p in self.lista_produtos:
            quantidade = int(input('Quantidade de {} desejada?'.format(p.nome)))
            valor = p.preco*quantidade
            print ("{}-{:.2f}€| Quantidade {} |{:.2f}€".format(p.nome,p.preco,quantidade,float(valor)))
            self.lista_quantidades.append(valor)
        return '\n'
            
            

    def total_do_pedido(self):
            conta = sum(self.lista_quantidades)
            return "Valor total das sua conta: {:.2F}€".format(conta)


    def __str__(self):
        return "Lista de compra criada com sucesso"


#Produtos
p1 = Produto(347865289, 'Chips',1.4,'Snack')
p2 = Produto(347865290, 'Cola-Cola',1.8,'Bebida')
p3 = Produto(347865287,'Rebucado',0.10,'Doces')
p4 = Produto(347865288,'Nesquik',2.80,'Cerais')

print(p1,p2,p3,p4)
print (p1.calculo_total())

#Pedidos
lc = Pedido()
lc.lista_produtos.append(p1)
lc.lista_produtos.append(p2)
lc.lista_produtos.append(p3)

print (lc)
print (lc.mostrar_pedido())
print (lc.total_do_pedido())
````

### O plano cartesiano

Representa um espaço bidimensional (em 2 dimensões), formado por duas linhas perpendiculares, uma horizontal e outra vertical, que se intersetam num ponto. A linha horizontal é chamada eixo das abcissas ou __eixo X__, enquanto a vertical é chamada eixo das ordenadas ou simplesmente o __eixo Y__.
Quanto ao ponto de interseção, é conhecido como o ponto de origem O. __O plano está dividido em 4 quadrantes.__


<img src="./Eixos.png"  />



### Pontos e coordenadas

O objetivo de tudo isto é descrever a posição dos **pontos** no plano sob a forma de **coordenadas**, que são formadas associando o valor do eixo X (horizontal) com o valor do eixo Y (vertical).


A representação de um ponto é simples: **P(X,Y)** onde X e Y são a distância horizontal (esquerda ou direita) e vertical (para cima ou para baixo) 
respetivamente, usando como referência o ponto de origem (0,0), mesmo no centro do plano.

<img src="./Pontos.png"  />


### Vetores no plano

Finalmente, um vetor no plano refere-se a um segmento orientado, gerado a partir de dois pontos diferentes.

Para efeitos práticos, é simplesmente uma linha formada a partir de um ponto inicial na direção de outro ponto final, pelo que se entende que um vetor tem comprimento e direção/sentido.

<img src="./Vetor.png"  />


Nesta figura, podemos ver dois pontos A e B que poderíamos definir como se segue:
* **A(x1, y1)** => **A(2, 3)**
* **B(x2, y2)** => **B(7, 5)**

E o vetor seria representado como a diferença entre as coordenadas do segundo ponto em relação ao primeiro (o segundo menos o primeiro):
* **AB = (x2-x1, y2-y1)** => **(7-2, 5-3)** => **(5,2)** 

O que, no final, é tudo o que é:

5 à direita e 2 no topo.
