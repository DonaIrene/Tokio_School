### O plano cartesiano

Representa um espaço bidimensional (em 2 dimensões), formado por duas linhas perpendiculares, uma horizontal e outra vertical, que se intersetam num ponto. A linha horizontal é chamada eixo das abcissas ou __eixo X__, enquanto a vertical é chamada eixo das ordenadas ou simplesmente o __eixo Y__.
Quanto ao ponto de interseção, é conhecido como o ponto de origem O. __O plano está dividido em 4 quadrantes.__


<img src="./Eixos.png"  />



### Pontos e coordenadas

O objetivo de tudo isto é descrever a posição dos **pontos** no plano sob a forma de **coordenadas**, que são formadas associando o valor do eixo X (horizontal) com o valor do eixo Y (vertical).


A representação de um ponto é simples: **P(X,Y)** onde X e Y são a distância horizontal (esquerda ou direita) e vertical (para cima ou para baixo) 
respetivamente, usando como referência o ponto de origem (0,0), mesmo no centro do plano.

<img src="./Pontos.png"  />


### Vetores no plano

Finalmente, um vetor no plano refere-se a um segmento orientado, gerado a partir de dois pontos diferentes.

Para efeitos práticos, é simplesmente uma linha formada a partir de um ponto inicial na direção de outro ponto final, pelo que se entende que um vetor tem comprimento e direção/sentido.

<img src="./Vetor.png"  />


Nesta figura, podemos ver dois pontos A e B que poderíamos definir como se segue:
* **A(x1, y1)** => **A(2, 3)**
* **B(x2, y2)** => **B(7, 5)**

E o vetor seria representado como a diferença entre as coordenadas do segundo ponto em relação ao primeiro (o segundo menos o primeiro):
* **AB = (x2-x1, y2-y1)** => **(7-2, 5-3)** => **(5,2)** 

O que, no final, é tudo o que é:

5 à direita e 2 no topo.

````python
import math
from cmath import sqrt

# Fazer o exercício aqui
#Criar uma classe chamada Ponto com as suas duas coordenadas X e Y.
class Ponto:
    __cordenadas = ''
    def __init__(self,X,Y):
        self.X = X
        self.Y = Y

        if self.X == "":
            self.X= 0
        else: pass

        if self.Y == "":
            self.Y = 0
        else: pass

        self.X = int(self.X)
        self.Y = int(self.Y)

        self.__cordenadas = X,Y

    @property
    def cordenadas(self):
        return self.__cordenadas

    @cordenadas.setter
    def cordenadas (self, novaCordenada):
        self.__cordenadas = novaCordenada

    #Substituir o método string ( __str__ ), para que ao imprimir um ponto por ecrã apareça em formato(X,Y)
    def __str__(self):
        return '({},{})'.format(self.X,self.Y) 

    #Adicionar um método construtor para criar pontos facilmente. Se não se recebe uma coordenada, o seu valor será zero
    def criar_ponto(self):
        self.X,self.Y =[i for i in input('Introduza ponto final X e Y separados por uma virgula:').split(",")]
        if self.X == "":self.X = 0
        if self.Y == "":self.Y = 0
        self.X = int(self.X)
        self.Y = int(self.Y)
        self.cordenadas = self.X,self.Y
        return f"ponto1: {self.X},{self.Y}"

    #Adicionar um método chamado quadrante que indique a quequadrante pertence o ponto,ou se é a origem
    def quadrante(self):

        if self.X == 0 and self.Y == 0:
            return "Origem"
        elif self.X > 0 and self.Y >= 0:
            return "Quadrante I"
        elif self.X > 0 and self.Y <= 0 :
            return "Quadrante II"
        elif self.X < 0 and self.Y <= 0:
            return "Quadrante III"
        elif self.X < 0  and self.Y >= 0:
            return "Quadrante IV"   

    #Agregar um método chamado vetor, que tome outro ponto e calcule o vetor resultante entre os dois pontos
    def vetor(self,ponto2):
        x2 = ponto2[0]
        y2 = ponto2[1]
        a = x2 - self.X
        b = y2 - self.Y
        c = sqrt(a+b)
        return "{:.02f}".format(c)

    #Incorporar um método chamado distância, que tome outro ponto e calcule a distância entre os dois pontos e a mostre por ecrã
    def distancia(self,ponto2):
        x2 = ponto2[0]
        y2 = ponto2[1]

        distancia = (x2-self.X)**2 + (y2-self.Y)**2

        return "{:.02f}".format(math.sqrt(distancia))

#Criar uma classe chamada Retângulo com dois pontos (inicial e final) que fazema diagonal do retângulo
class Retangulo(Ponto):
    def __init__(self,inicial=[],final=[]):
        
        self.x1 = inicial[0]
        self.y1 = inicial[1]
        self.x2 = final[0]
        self.y2 = final[1]
        self.inicial = inicial
        self.final = final

    #Adicionar ao retângulo um método chamado base que mostre a base.
    def mostrar_base(self):
        if self.inicial[0] < 0: self.inicial[0] -= self.inicial[0]*2
        if self.final[0] < 0: self.final[0] -= self.final[0]*2
        base = abs(self.inicial[0] + self.final[0])
        return f"O retangulo tem {base} px de base"
    
    #Agregar ao retângulo um método chamado altura que mostre a altura
    def mostrar_altura(self):
        if self.inicial[1] < 0: self.inicial[1] -= self.inicial[1]*2
        if self.final[1] < 0: self.final[1] -= self.final[1]*2

        altura = self.inicial[1] + self.final[1]
        return f"O retangulo tem {altura} px de altura"
    
    #Adicionar ao retângulo um método chamado área que mostre a área
    def mostrar_area(self):
        if self.inicial[0] < 0: self.inicial[0] -= self.inicial[0]*2
        if self.final[0] < 0: self.final[0] -= self.final[0]*2
        if self.inicial[1] < 0: self.inicial[1] -= self.inicial[1]*2
        if self.final[1] < 0: self.final[1] -= self.final[1]*2
        
        area = abs(self.inicial[0]+self.final[0]) * abs(self.inicial[1]+self.final[1])
        return f"O retangulo tem {area} px de area"
    
    #Incorporar um método construtor para criar ambos os pontos facilmente, se não se enviam criam-se dois pontos na origem por defeito
    def inserir_pontos(self):
        self.x1,self.y1 =[i for i in input('Introduza ponto final X e Y separados por uma virgula:').split(",")]
        self.x2,self.y2 =[i for i in input('Introduza ponto final X e Y separados por uma virgula:').split(",")]
        if self.x1 == "":self.x1 = 0
        if self.y1 == "":self.y1 = 0
        if self.x2 == "":self.x2 = 0
        if self.y2 == "":self.y2 = 0
        self.inicial[0] = int(self.x1)
        self.inicial[1] = int(self.y1)
        self.final[0] = int(self.x2)
        self.final[1] = int(self.y2)
        return f"ponto1: {self.inicial[0]},{self.inicial[1]} || ponto2: {self.x2},{self.y2}"
    
    
    def __str__(self):
        return f"{self.inicial} || {self.final}"
    
## Criar os pontos A(2, 3), B(5,5), C(-3, -1) e D(0,0) e imprimir por ecrã##
A = Ponto(2 ,3)
B = Ponto (5 ,5)
C = Ponto(-3, -1)
D = Ponto(0 ,0)
print(f' A = {A.cordenadas} \n B = {B.cordenadas} \n C = {C.cordenadas} \n D = {D.cordenadas}')

#Averiguar a que quadrante pertence o ponto A, C e D
print(f' A - {A.quadrante()} \n C - {C.quadrante()} \n D - {D.quadrante()}')

#Consultar os vetores AB e BA#
print(f' Vetor AB - {A.vetor(B.cordenadas)}')
print(f' Vetor BA - {B.vetor(A.cordenadas)}')

#Consultar a distância entre os pontos 'A e B' e 'B e A'
print(f"A distancia de A para B e de  : {A.distancia(B.cordenadas)}")
print(f"A distancia de B para A e de : {B.distancia(A.cordenadas)}")

#Determinar qual dos 3 pontos A, B ou C, se encontra mais longe da origem, ponto (0,0)
if B.distancia(D.cordenadas) > A.distancia(D.cordenadas) and B.distancia(D.cordenadas) > C.distancia(D.cordenadas):
        print("O ponto mais distante da origem e o ponto  B.")
if A.distancia(D.cordenadas) > B.distancia(D.cordenadas) and A.distancia(D.cordenadas) > C.distancia(D.cordenadas):
        print("O ponto mais distante da origem e o ponto  A.")
if C.distancia(D.cordenadas) > B.distancia(D.cordenadas) and C.distancia(D.cordenadas) > A.distancia(D.cordenadas):
        print("O ponto mais distante da origem e o ponto  C.")

# Fazer um retângulo utilizando os pontos A e B
retangulo1 = Retangulo(A.cordenadas,B.cordenadas)

#Consultar a base, altura e área do retângulo
print(retangulo1.mostrar_base())
print(retangulo1.mostrar_altura())
print(retangulo1.mostrar_area())
````
