```python
class Filme:
    def __init__(self, titulo, durancao, lancamento):
        self.titulo=titulo
        self.duracao = durancao
        self.lancamento = lancamento
        print("Criacao do filme:",titulo)

    def __str__(self):
        return "{} ({})".format(self.titulo,self.lancamento)

class Catalogo:
    filmes = []
    def __init__(self,filmes = []):
        self.filmes=filmes
        print ("Criacao de um catalogo")

    def agregar(self,p):
        self.filmes.append(p)
        print("Filmes {}, foi adicionado a lista".format(p.titulo))

    def mostrar(self):
        for p in self.filmes:
            print(p)

d = Filme("O Padrinho",175,1972)
c = Catalogo([d])
print("\n")

c.mostrar()
c.agregar(Filme("O Padrinho: part 2",202,1974))
print("\n")

c.mostrar()
print("\n")

c2 = Catalogo()
c2.mostrar()
print("\n")

pM1 = Filme("Matrix", 175,1999)
pM2 = Filme("Matrix Reloaded",175,2003)
pM3 = Filme("Matrix Revolution",175,2003)
print("\n")

c3 = Catalogo([pM1,pM2,pM3])
c3.mostrar()
```
